```python
import qoery
```

```python
client = qoery.Client()
```

```python
candles = client.candles.get(
        symbol="WETH-USDC", 
        interval="15m", 
        limit=100
    )
candles.df
```

```python
ticks = client.ticks.get(
            symbol="WETH-USDC", 
            limit=100)
ticks.df
```
